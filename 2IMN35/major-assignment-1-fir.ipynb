# Major assignment *Symmetric FIR-filter Design* (d.d. 17/5/2021)

In this assignment you will design, implement and analyse a fine-grained symmetric FIR-filter of order $2N{+}1$. So, we consider a filter with an odd number of coefficients (taps). 

For $0 \leq N$, the specification of such a filter is given by the i/o-relation:

$$
\begin{eqnarray*}
B(i) & = & (\Sigma j: 0 \leq j < N: H(j){\cdot}(A(i{-}j) + A(i{-}2N{+}j))\;) + H(N){\cdot}A(i{-}N), \qquad\mathrm{~for~}2N \leq i \\
\end{eqnarray*}
$$

where the set of filter coefficients $\{ H(j) \mid 0 \leq j \leq 2N\}$ satisfies $H(j) = H(2N{-}j)$, for $0 \leq j \leq N$. 

Note that the specification exploits the symmetry of the coefficients by  using only  $\{ H(j) \mid 0 \leq j \leq N\}$.

# Fine-grained design

The next thing you have to do is 

- <span style="color:blue"><b>
    give a formal derivation of a fine-grained filter,
</b></span> 

i.e., a design in which each filter stage handles the contribution of at most one filter coeficient. The result should be a **<span style="color:blue">RIA</span>**, i.e., a set of i/o-relations  that, for each stage, relate its output streams to its input streams.

Of course, there are multiple designs possible, and you are <span style="color:green">free to make your own one</span>, but to help you a little bit, you may strive for a solution where the generic stage is equipped with an output port that produces an output stream satisfying 

$$
\begin{eqnarray*}
b[n](i) & = & (\Sigma j: n \leq j < N: H(j){\cdot}(A(i{-}j) + A(i{-}2N{+}j))\;) + H(N){\cdot}A(i{-}N), 
\end{eqnarray*}
$$

and with input ports consuming streams satisfying

$$
\begin{eqnarray*}
a[n](i) & = & A(i{-}n)      \\
c[n{+}1](i) & = & A(i{-}2N{+}n) \\
\end{eqnarray*}
$$

Note that all bounds for indices $i$ and $n$ are missing, as is the precise number of stages and their nature (generic/boundary), and that the system input $A$ still occurs in the i/o-relations.
<span style="color:blue">It is up to you to complete the design by removing these shortcomings.</span>
Think carefully about the direction of the streams and deviating behavior of boundary cells! Where relevant motivate your design decisions.

Before derivation, we first observe the original specification. Since the set of filter coefficients $\{H(j)\ |\  0 \le j \le 2N\}$ satisfies $H(j)=H(2N-j)$, for $0 \le j \le 2N$,

$$
\begin{align}
B(i)
& = \sum_{0 \le j < N}[H(j)(A(i-j)+A(i-2N+j))] + H(N)A(i-N) \\
& = \sum_{0 \le j < N}[H(j)A(i-j) + H(j)A(i-2N+j)] + H(N)A(i-N) \\
& = \sum_{0 \le j < N}[H(j)A(i-j) + H(2N-j)A(i-(2N-j))] + H(N)A(i-N) \\
& = \sum_{0 \le j < N}H(j)A(i-j) + \sum_{0 \le j < N}H(2N-j)A(i-(2N-j)) + H(N)A(i-N)
\end{align}
$$

Now we assume that $k=2N-j$, since $0\le j < N$, we have $N<t\le 2N$. All $2N-j$ will be substituded by $t$ in the second term of the above formula, then

$$
\begin{align}
B(i)
& = \sum_{0 \le j < N}H(j)A(i-j) + \sum_{N < k \le 2N}H(k)A(i-k) + H(N)A(i-N) \\
& = \sum_{0 \le j < N}H(j)A(i-j) + \sum_{N \le k \le 2N}H(k)A(i-k) \\ 
\end{align}
$$

We can find that the form of these two terms are exactly the same, so we can **re-write this specification using a new variable $t$ and $N^{\star}$, where $0 \le t <N^{\star}$ and $N^{\star}=2N+1$**.

$$
B(i) = \sum_{0 \le t < N^{\star}}H(t)A(i-t) 
$$

Next, we can perform the derivation. First, this FIR system is designed as a set of stages, *i.e.* $FIR=\{fir_n\ |\ 0 \le n \le N^{\star}\}$. The output stream from the output port $b[n]$ of an equip stage $fir_n$ is

$$
\begin{align}
b[n](i)
& = \sum_{0 \le t < n}h(t)A(i-t) \\
& = \sum_{0 \le t < n-1}h(t)A(i-t) + h(n-1)A(i-n+1) \\
& = b[n-1](i) + h(n-1)A(i-n+1)
\end{align}
$$

where $n-1 \le i$. Then, similar to the video lecture, 

$$
a[n-1](i) = A(i-n+1), \ n-1 \le i
$$
and
$$
a[n](i) = a[n-1](i-1), \ 1 \le n < N^{\star},\ n \le i
$$

In this way, we can get the regular iterative algorithm (RIA) result.

$$
\begin{array}[t]{rclcr}
b[0](i) & = & 0,                               &                          & 0 \leq i \\
\\
a[n](0) & = & \Box,                            & 1 \leq n < N^{\star},    &          \\
a[n](i) & = & a[n{-}1](i{-}1),                 & 1 \leq n < N^{\star},    & 1 \leq i \\
b[n](i) & = & b[n{-}1](i) + {}                                           \\
        &   & h(n{-}1){{\cdot}} a[n{-}1](i),   & 1 \leq n \leq N^{\star}, & 0 \leq i \\
%&&&\\
\hline
a[0](i) & = & A(i),    & & 0 \leq i \\
B(i)    & = & b[N^{\star}](i), & & 0 \leq i \\
\end{array}
$$

where $\Box$ denotes an arbitrary value (start-up noise).

We need to make $i$ as the firing counter. Here we use out-aligned / production-based style. Then we can re-write the formulas.

$$
\begin{array}[t]{rclcr}
b[0](i) & = & 0,                               &                  & 0 \leq i \\
\\

a[n{-}1](-1) & = & \Box, & 1 \leq n < N^{\star} \\
\end{array}
$$

```python
# Stage definition production-based (output-aligned)
class FIR_n_pb(Node):
    def __init__(self, I=[], h=0):
        super(FIR_n_pb, self).__init__(I=I)
        # fo[0] = a[n](i)        , fo[1] = b[n](i), 
        #  x[0] = a[n-1](i-1:i+1), x[1] = b[n-1](i)
        self.fo = [lambda x: x[0][0], lambda x: x[1] + h*x[0][1]]
        
    def init(self):         
        box = 0                            # arbitrary value chosen 0
        self.I[0].init(D=0, S=1, x=[box])  # first a-input arbitrarily 0
```

Now we specify some hyper-parameters.

```python
N_star = 2*N_audio+1
h = coef
```

```python
G_sfir_fg = Graph(rate=44100); 
G = G_sfir_fg 
G.fir = [Node(fo= lambda x: 0)]+\
        [FIR_n_pb(h=h[n-1]) for n in range(1,N_star+1)]
G.mp3 = MP3(mp3='../audio/sultans.mp3', tmin=0.7, rate=44100)
G.snk_a, G.B = Node(), Node()
G.a = Path([G.mp3] + G.fir[1:] + [G.snk_a])
G.b = Path(G.fir + [G.B])
G.build()
G.plot_graph()
```

## Redesign the filter

We first modify the coefficients of the FIR filter to type `np.int16`.

```python
temp = (2**15)*coef_audio
new_coef = temp.astype(np.int16)

plt.plot(new_coef)
plt.plot(new_coef, 'bo')
plt.show()
```

Next, class `FIR_n_pb` also need to be modified accordingly. So we define a new version `N_FIR_n_pb` as follows.

```python
# Stage definition production-based (output-aligned)
class N_FIR_n_pb(Node):
    def __init__(self, I=[], h=0):
        super(N_FIR_n_pb, self).__init__(I=I)
        self.fo = [lambda x: x[0][0], lambda x: np.int16(x[1] + h*np.int64(x[0][1])//(2**15))]
        
    def init(self):         
        box = 0                            # arbitrary value chosen 0
        self.I[0].init(D=0, S=1, x=[box])  # first a-input arbitrarily 0
```

Next, we apply the new coefficients and define a new graph `G_N_sfir_fg`

```python
h = new_coef
```

```python
G_N_sfir_fg = Graph(rate=44100); 
G = G_N_sfir_fg 
G.fir = [Node(fo= lambda x: 0)]+\
        [N_FIR_n_pb(h=h[n-1]) for n in range(1,N_star+1)]
G.mp3 = MP3(mp3='../audio/sultans.mp3', tmin=0.7, rate=44100)
G.snk_a, G.B = Node(), Node()
G.a = Path([G.mp3] + G.fir[1:] + [G.snk_a])
G.b = Path(G.fir + [G.B])
G.build()
G.plot_graph()
```
