# A poor man's scaler

Inserts after each pair of values the average of the pair.

```python
def UP(A):
#    A = np.uint16(A)   # to prevent overflow
    rdimA, cdimA = A.shape
    B = np.zeros((rdimA, (cdimA//2)*3))
    for i in range(cdimA//2):
        B[:,3*i:3*i+2] = A[:,2*i:2*i+2]
        B[:,3*i+2]     = np.uint8((A[:,2*i] + A[:,2*i+1])/2)
    return B    
```

```python
A   = image2array('../images/valve.png')
plt.imshow(A)
```

```python
B = UP(A)
C = np.transpose(UP(np.transpose(B)))
plt.imshow(C)
```
